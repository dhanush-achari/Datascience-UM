```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as ss

import seaborn as sns

%matplotlib notebook
%matplotlib inline


#happines data of different countries based on region and happiness scores in year 2016 nd 2020
df = pd.read_csv('2016.csv')

df2 = pd.read_csv('2020.csv')
plt.style.use('ggplot')
df2






```

```python
df.corr()
```

```python
sns.heatmap(df.corr())
```

```python
df.plot.hist('Happiness Score',bins=5)
```

```python
#grouping dataframes based on region and finding average happiness scores
avgdf2016=df.groupby('Region').agg({'Happiness Score':np.mean,'Health (Life Expectancy)':np.mean})
avgdf2020=df2.groupby('Regional indicator').agg({'Ladder score':np.mean,'Healthy life expectancy':np.mean})
avgdf2020.rename(columns={'Ladder score':'Happiness Score'},inplace=True)
list(avgdf2016.index)
avgdf2016.reset_index(inplace=True)
avgdf2016.Region=avgdf2016.Region.map({'Australia and New Zealand':'Australia and New Zealand',
 'Central and Eastern Europe':'Central and Eastern Europe',
 'Eastern Asia':'East Asia',
 'Latin America and Caribbean':'Latin America and Caribbean',
 'Middle East and Northern Africa':'Middle East and North Africa',
 'North America':'North America and ANZ',
 'Southeastern Asia':'Southeast Asia',
 'Southern Asia':'South Asia',
 'Sub-Saharan Africa':'Sub-Saharan Africa',
 'Western Europe':'Western Europe'})
avgdf2016.set_index('Region',inplace=True)
avgdf2020
avgdf=pd.merge(avgdf2016,avgdf2020,how='outer',left_index=True,right_index=True)
avgdf
avgdf.fillna(method='ffill', axis=1,inplace=True)
avgdf.fillna(method='bfill',axis=1,inplace=True)
avgdf
avgdf['avg']=avgdf.apply(lambda x:np.mean([x['Happiness Score_x'],x['Happiness Score_y']]),axis=1)
list(avgdf.index)
```

```python
plt.figure()
l=np.array(avgdf['Happiness Score_x'])
plt.plot(l,'ro-',label='Avg2016')
plt.plot(np.array(avgdf['Happiness Score_y']),'--b',label='Avg2020')
plt.plot(np.array(avgdf['avg']),'go--',label='Avgboth')

plt.xticks(np.arange(len(avgdf)),['Australia and New Zealand',
 'Central and Eastern Europe',
 'Commonwealth of Independent States',
 'East Asia',
 'Latin America and Caribbean',
 'Middle East and North Africa',
 'North America and ANZ',
 'South Asia',
 'Southeast Asia',
 'Sub-Saharan Africa',
 'Western Europe'],rotation=90)
plt.legend()
plt.ylabel('Happiness Score')
plt.xlabel('Region')
plt.title('Happiness 2016 vs 2020')
```
